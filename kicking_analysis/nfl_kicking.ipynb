```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load data
##### Source: Pro Football Reference, https://www.pro-football-reference.com/years/NFL/kicking.htm

```python
# data in kicking_data.csv
df = pd.read_csv('kicking_data.csv')

print(df.head())
```

# Find FG% by Year

```python
# create graph: FG% by year

year = df['Year']
ave = df['Total FGM'] / df['Total FGA'] * 100

fig, ax = plt.subplots()
ax.plot(year, ave, color='purple')
ax.set_title('NFL FG%\n 1960 - 2024')
ax.set_xlabel('Year')
ax.set_ylabel('Field Goal %')
ax.set_facecolor('darkgreen')
fig.set_facecolor('gray')
#plt.savefig('NFL_FG_pct.png', bbox_inches = 'tight')
plt.show()
```

```python
# create graph: Field Goals Attempted Vs Made by Year

year = df['Year']
FGM = df['Total FGM']
FGA = df['Total FGA']

fig, ax = plt.subplots()
ax.plot(year, FGA, color='blue')
ax.plot(year, FGM, color='orange')
ax.set_xlabel('Year')
ax.set_ylabel('Field Goals Kicked')
ax.set_title('NFL FG Attempted vs Made\n from 1960-2024')
ax.set_facecolor('darkgreen')
fig.set_facecolor('grey')
ax.legend(['Attempted', 'Made'], loc='lower right', fontsize='large', title='Field Goals')
#plt.savefig('NFL_FG_Attempted_Made.png', bbox_inches = 'tight')
plt.show()
```

```python
# Why the big increase in 1970?
## A: added teams = more total field goals
# Why the attempts drop off in 1974?
## A: moved the goal posts to the back of the endzone
teams = df[['Year', 'Teams', 'Total FGA']]
print(teams.head(20))
```

```python
# create graph: FG% by Distance

year = df['Year']
total_ave = df['Total FGM'] / df['Total FGA'] * 100
ave019 = df['0-19 FGM'] / df['0-19 FGA'] * 100
ave2029 = df['20-29 FGM'] / df['20-29 FGA'] * 100
ave3039 = df['30-39 FGM'] / df['30-39 FGA'] * 100
ave4049 = df['40-49 FGM'] / df['40-49 FGA'] * 100
ave50 = df['50+ FGM'] / df['50+ FGA'] * 100

fig, ax = plt.subplots()
ax.plot(year, total_ave, color='purple')
ax.plot(year, ave019)
ax.plot(year, ave2029)
ax.plot(year, ave3039)
ax.plot(year, ave4049)
ax.plot(year, ave50)
ax.set_title('NFL FG% by Distance\n 1960 - 2024')
ax.set_xlabel('Year')
ax.set_ylabel('Field Goal %')
ax.legend(['Total Ave', '0-19', '20-29', '30-39', '40-49', '50+'], title='Yardage')
ax.set_facecolor('darkgreen')
fig.set_facecolor('gray')
#plt.savefig('NFL_FG_pct_Distance.png', bbox_inches = 'tight')
plt.show()
```

```python
# create graph: Field Goals Attempted by Distance

year = df['Year']
total_ave = df['Total FGA']
att019 = df['0-19 FGA']
att2029 = df['20-29 FGA']
att3039 = df['30-39 FGA']
att4049 = df['40-49 FGA']
att50 = df['50+ FGA']

fig, ax = plt.subplots()
#ax.plot(year, total_ave, color='purple')
ax.plot(year, att019)
ax.plot(year, att2029)
ax.plot(year, att3039)
ax.plot(year, att4049)
ax.plot(year, att50)
ax.set_title('NFL FGA by Distance\n 1960 - 2024')
ax.set_xlabel('Year')
ax.set_ylabel('Field Goals Attempted')
ax.legend(['0-19', '20-29', '30-39', '40-49', '50+'], title='Yardage')
ax.set_facecolor('darkgreen')
fig.set_facecolor('gray')
#plt.savefig('NFL_FG_attempted_Distance.png', bbox_inches = 'tight')
plt.show()
```
